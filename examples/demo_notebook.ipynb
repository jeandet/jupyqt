# jupyqt Demo

This notebook demonstrates interacting with the host PySide6 application from JupyterLab cells.

## Exposed objects

The host app pushes several objects into the kernel namespace:
- `window` — the main QMainWindow (via QtProxy)
- `counter_label` — a QLabel showing a counter
- `increment()` — bump the counter
- `reset()` — reset the counter to zero

```python
# Read the current counter value
counter_label.text()
```

```python
# Increment the counter a few times
for _ in range(5):
    increment()
counter_label.text()
```

```python
# Change the window title from the notebook
window.setWindowTitle("Controlled from Jupyter!")
```

```python
# Reset the counter
reset()
counter_label.text()
```

## Async support

Top-level `await` works in cells.

```python
import asyncio

for i in range(3):
    await asyncio.sleep(0.5)
    increment()
    print(f"step {i+1}: {counter_label.text()}")
```
