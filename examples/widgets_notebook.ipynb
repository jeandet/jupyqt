# ipywidgets in jupyqt

Interactive widgets work via the comm protocol.
Requires: `pip install ipywidgets`

```python
import ipywidgets as widgets

slider = widgets.IntSlider(value=50, min=0, max=100, description="Value:")
display(slider)
```

```python
# Read the slider value (update after dragging)
slider.value
```

```python
text = widgets.Text(description="Name:")
greeting = widgets.HTML(value="<em>Type your name above</em>")

def on_change(change):
    greeting.value = f"<h3>Hello, {change['new']}!</h3>"

text.observe(on_change, names="value")
display(widgets.VBox([text, greeting]))
```

```python
import numpy as np
from ipywidgets import interact

@interact(frequency=(1, 10, 0.5), amplitude=(0.1, 2.0, 0.1))
def plot_wave(frequency=3, amplitude=1.0):
    import matplotlib.pyplot as plt
    x = np.linspace(0, 2 * np.pi, 200)
    plt.figure(figsize=(8, 3))
    plt.plot(x, amplitude * np.sin(frequency * x))
    plt.ylim(-2.5, 2.5)
    plt.title(f"f={frequency} Hz, A={amplitude}")
    plt.show()
```
